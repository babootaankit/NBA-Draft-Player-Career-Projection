```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_source = "Resources/historical_projections.csv"
draft_data = pd.read_csv(data_source)
draft_data.head(15)
```

```python
data_draft_source = "Resources/draftees.csv"
draftees_data = pd.read_csv(data_draft_source)
draftees_data = draftees_data.loc[draftees_data["Draft_Yr"] >= 2001]
draftees_data
```

```python
Player = draftees_data["Player"]
Draft_pick = draftees_data["Pk"]
pick = pd.DataFrame({"Player": Player, "Pick": Draft_pick})
pick
```

```python
combined_data = pd.merge(pick, draft_data, on="Player")
combined_data
```

```python
starter = plt.scatter(combined_data["Pick"], combined_data["Starter"])
plt.xlabel("Place Drafted")
plt.ylabel("Probability of becoming a starter")
plt.title("Chances of becoming a starter")
plt.savefig("Visualizations/starter.png")
```

```python
role = plt.scatter(combined_data["Pick"], combined_data["Role Player"])
plt.xlabel("Place Drafted")
plt.ylabel("Probability of becoming a role player")
plt.title("Chances of becoming a role player")
plt.savefig("Visualizations/role.png")
```

```python
bust = plt.scatter(combined_data["Pick"], combined_data["Bust"])
plt.xlabel("Place Drafted")
plt.ylabel("Probability of becoming a bust")
plt.title("Chances of becoming a bust")
plt.savefig("Visualizations/bust.png")
```

```python
superstar = plt.scatter(combined_data["Pick"], combined_data["Superstar"])
plt.xlabel("Place Drafted")
plt.ylabel("Probability of becoming a superstar")
plt.title("Chances of becoming a superstar")
plt.savefig("Visualizations/superstar.png")
```
